```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv("car_price.csv")
```

```python
raw_data.head(10)
```

## Issues to handle

1. Seperating Brand and Model from brand column
2. Seperating year from brand column
3. Seperating Kms, fuel from features column
4. Removing 'kms', and comma from kms column
5. Removing rupees sign and comma from price column
6. Changing datatypes of few columns

```python
raw_data[['Model','Year']] = raw_data.brand.str.split("-",expand=True,)
```

```python
raw_data.head()
```

```python
raw_data["Brand"]=raw_data['Model'].str.split(' ').str[0]
```

```python
raw_data["Brand"] = raw_data['Brand'].replace({'Land': 'Land Rover', 'Mini': 'Mini Cooper'})
```

```python
raw_data.head()
```

```python
raw_data.drop("brand", axis=1, inplace =True)
```

```python
raw_data.head()
```

```python
raw_data['Kms_driven'] = raw_data.features.str.split("/").str.get(0)
raw_data['Fuel_type'] = raw_data.features.str.split("/").str.get(1)
```

```python
raw_data.head()
```

```python
raw_data.drop("features", axis=1, inplace =True)
```

```python
raw_data.head()
```

```python
raw_data.Brand.unique()
```

```python
raw_data['Kms_driven']=raw_data['Kms_driven'].str.split().str.get(0).str.replace(',','').astype(np.int64)
```

```python
raw_data.head()
```

```python
#Price column has "Ask For Price", removing it
raw_data=raw_data[raw_data['price']!='Ask For Price']
```

```python
raw_data["Price"] = raw_data.price.str.split("₹",expand=True,).get(1).str.replace(',','').astype(np.int64)
```

```python
raw_data.drop("price",axis=1,inplace=True)
```

```python
raw_data.head()
```

```python
raw_data['Model']=raw_data['Model'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
raw_data.head()
```

```python
print(raw_data.dtypes)
```

```python
raw_data["Year"]=raw_data.Year.astype(np.int64)
```

```python
print(raw_data.dtypes)
```

```python
raw_data.isnull().sum()
```

```python
raw_data = raw_data.dropna()
```

```python
raw_data.isnull().sum()
```

```python
raw_data.to_csv('Cleaned_car_data.csv',index=False)  
```
